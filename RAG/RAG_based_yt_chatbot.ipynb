```python
from youtube_transcript_api import YouTubeTranscriptApi
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores.faiss import FAISS
from langchain_core.prompts import PromptTemplate
```

# Step 1: Ingestion (Indexing)

```python
!pip install --no-cache-dir youtube-transcript-api
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
from langchain_core.documents import Document

# Set your YouTube Video ID
video_id = "Gfr50f6ZBvo"  # Replace with any video ID you want

# Fetch the transcript and convert to plain text
try:
    ytt_api = YouTubeTranscriptApi()
    # Specify the language (e.g., 'en' for English)
    transcript = ytt_api.fetch(video_id, languages=['en'])

    # Join all transcript pieces into one full string
    transcript_text = "\n".join([item.text for item in transcript.snippets])

    print("Transcript fetched successfully!\n")
    print(transcript_text)  # Print first 1000 characters

except Exception as e:
    print("Transcript not available:", e)
    transcript_text = ""
```

# Step 1b:  Indexing(Text Splitting)

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)

chunks = splitter.create_documents([transcript_text])
```

```python
len(chunks)
```

```python
chunks[78]
```

## Step 1c & 1d - Indexing (Embedding Generation and Storing in Vector Store)


```python
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/embedding-001",
    google_api_key=api
)

vectorstore = FAISS.from_documents(chunks, embeddings)
```

```python
vectorstore.index_to_docstore_id
```

```python
vectorstore.get_by_ids(['b0dcbdfb-5e30-4372-8ce1-85308f339bc7'])
```

# Step2: Retriver

```python
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 5}
)
```

```python
retriever
```

```python
retriever.invoke('what is deepmind')
```

# step3: Augumentation

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.2,
    google_api_key=api
)
```

```python
prompt = PromptTemplate(
    template="""
    You are a helpful assistant.
    Answer only from the provided transcript context.
    If the context is insufficient, just say you don't know.
    {context}
    Question: {question}
    """,
    input_variables=["context", "question"]
)
```

```python
question = 'is the topic of aliens discussed in this video? If so than what was discussed'
retrived_docs = retriever.invoke(question)
```

```python
context_text = "\n\n".join([doc.page_content for doc in retrived_docs])
```

```python
retrived_docs
```

```python
context_text
```

```python
final_prompt = prompt.invoke(
    {
        "context": context_text,
        "question": question
    }
)
```

```python
final_prompt
```

# step4: Generation

```python
answer = llm.invoke(final_prompt)
print(answer.content)
```

# Building a Chain

```python
from langchain_core.runnables import RunnableParallel, RunnablePassthrough, RunnableLambda
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(retrived_docs):
  context_text = " ".join([doc.page_content for doc in retrived_docs])
  return context_text
```

```python
parallel_chain = RunnableParallel(
   context=retriever | RunnableLambda(format_docs),
   question=RunnablePassthrough()
)
```

```python
parallel_chain.invoke('what is deepmind')
```

```python
parser = StrOutputParser()
```

```python
main_chain = parallel_chain | prompt | llm | parser
```

```python
main_chain.invoke('Can you summarize this video?')
```
